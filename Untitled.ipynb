```python
import os.path as osp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def get_acc(df):
    # DataFrame olumns are
    # Train_Loss
    # Train Accuracy
    # Validation Accuracy
    # Test Accuracy
    best_val_acc=0
    test_acc=0
    train_acc=0
    for i in range(len(df)):
        if df["Validation Accuracy"].iloc[i] > best_val_acc:
            best_val_acc = df["Validation Accuracy"].iloc[i]
            test_acc = df["Test Accuracy"].iloc[i]
            train_acc = df["Test Accuracy"].iloc[i]
    return train_acc, test_acc, best_val_acc, df["Test Accuracy"].iloc[-1]
```

```python
# Need to plot best accuracies of all datasets and
# the network commfiguration which achieved so.
# Given a dataset, the networt must return the best
# accuracy along with the network hyperparameters

def get_best_acc(dataset):
    h_128 = []; b_128 = []; test_128 = []; val_128 = []; train_128 = []; train_best_128=[];
    for h in [1, 2, 4, 8]:
        for b in [1, 2, 4, 8]:
            h_128 += [h] ;  
            b_128 += [b] ;
            #name = dataset + '_' + b + '_' + h + '_' + emb + '_' + alpha + '1.csv'
            df = pd.read_csv("./Performance/"+dataset + "/"+str(b)+"_"+str(h)+"_128_0.csv") ;
            train_acc, test_acc, val_acc, train_best_acc = get_acc(df) ;
            test_128 += [test_acc] ;
            val_128 += [val_acc] ;
            train_128 += [train_acc]
            train_best_128 += [train_best_acc]
    DF = pd.DataFrame({
                        "num_heads": h_128,
                        "num_blocks": b_128,
                        "train_acc_128": train_128,
                        "test_acc_128": test_128,
                        "val_acc_128": val_128,
                        "train_best_acc_128": train_best_128
                        })
    print("Accuracy 128:",max(test_128))
    index_128 = test_128.index(max(test_128))
    print("128",h_128[index_128], b_128[index_128])
    return DF
```

```python
dt = get_best_acc("NCI1")
```

```python
plt.plot(dt[dt["num_heads"]==1].num_blocks, dt[dt["num_heads"]==1].test_acc_128, "-r", label="Heads=1")
plt.plot(dt[dt["num_heads"]==2].num_blocks, dt[dt["num_heads"]==2].test_acc_128, "-g", label="Heads=2")
plt.plot(dt[dt["num_heads"]==4].num_blocks, dt[dt["num_heads"]==4].test_acc_128, "-b", label="Heads=4")
plt.plot(dt[dt["num_heads"]==8].num_blocks, dt[dt["num_heads"]==8].test_acc_128, "-y", label="Heads=8")
#plt.plot(buff.num_blocks, buff.val_acc_64, "-g", label="Val64")
#plt.plot(buff.num_blocks, buff.val_acc_128, "-y", label="Val128")
plt.legend(loc="upper left")
plt.ylim(0.5, 0.8)
plt.ylabel("Accuracy")
plt.xlabel("Number of Blocks")
plt.title("Test Accuracy vs. Num_blocks (emb.=128)- NCI1")
plt.show()

#enter image description here
```
